````python
import tensorflow as tf
from Main_discriminator import inference
from Main_deeper_u_net import generator
import numpy as np
import hdf5storage
import scipy.io as sio
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import cv2
from tensorflow.keras import Model


def normalize_0_to_1(mat):
    for i in range(mat.shape[0]):
        mat[i]=(mat[i]-mat[i].min())/(mat[i].max()-mat[i].min())
    return mat

def custom_loss(weight):
    def mse_loss(y_ground_truth,y_output):
        return weight*tf.reduce_mean(tf.square(tf.subtract(y_ground_truth, y_output)))
        #return weight*tf.reduce_mean(tf.square(tf.linalg.norm(tf.subtract(y_ground_truth, y_output), ord='fro', axis = [1, 2])))
    return mse_loss



def get_training_batch(training_data_batch,masks):
    return np.sum(training_data_batch * masks, axis=3)[:, :, :, np.newaxis]

# Define gan
def get_gan(gen, dis, optimizer):
    dis.trainable = True
    gen_input = gen.input
    gen_output6, gen_output12, gen_output = gen.output
    gan_output = dis(gen_output)
    model = tf.keras.Model(gen_input, [gan_output, gen_output, gen_output12, gen_output6])
    losses = {
                "discriminator_lambda": tf.keras.losses.BinaryCrossentropy(from_logits=True),
                "out24": custom_loss(1),
                "out12": custom_loss(0.1),
                "out6": custom_loss(0.1),
            }
    lossWeights = {
                "discriminator_lambda": 1.0,
                "out24": 100.0,
                "out12": 100.0,
                "out6": 100.0,
            }
    model.compile(loss=losses, loss_weights=lossWeights, optimizer=optimizer)
    return model

optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4)

gen = generator(size=[200, 200, 100], training=True, weigh_decay=1e-8, output_bands=100, bands=100, upsampling = 1, Factor = 1e-4)

dis = inference(size=[200, 200, 100], training=True, weigh_decay=0.0)

gen.load_weights("checkpoints/genw_50L.tf")

model = get_gan(gen, dis, optimizer)

##
L = 100
M = 200
N = 200
path = '/content/gdrive/My Drive/Canada/Pr1_256/CNN_RealMea/Real_CA_V1.mat'
mask = hdf5storage.loadmat(path)['CA']
mask = np.float16(mask)
mask = np.divide(mask,tf.math.reduce_max(mask))
mask1 = mask
mask = tf.expand_dims(mask,axis=0)
mask = tf.expand_dims(mask,axis=-1)
#
img=hdf5storage.loadmat('/content/gdrive/My Drive/Canada/training_data_200x200x100/0.mat')['data'] #(10,256,256,24)
img = tf.expand_dims(img,axis=0)
img = np.divide(img,tf.math.reduce_max(img))


Aux1 = tf.multiply(mask, img)
Aux1 = tf.pad(Aux1, [[0, 0], [0, 0], [0, L - 1], [0, 0]])
Yg = None
for i in range(0,L):
    if Yg is not None:
      Yg = Yg + tf.roll(Aux1[:,:,:,i], shift=i, axis=2)
    else:
      Yg = Aux1[:,:,:,i]             
X = None
for i in range(0,L):
  if X is not None:
    Ax = tf.expand_dims(tf.multiply(mask1, Yg[:, :, i:N+i]), -1)
    X = tf.concat([X, Ax], axis=-1)    
  else:
    X = tf.expand_dims(tf.multiply(mask1,Yg[:, :, 0:N]), -1)   
Yg = tf.expand_dims(Yg,axis=-1)
Yg = np.divide(Yg,tf.math.reduce_max(Yg))
X = np.float16(np.divide(X,tf.math.reduce_max(X)))
g12 = []
_, _, g12 = gen.predict(X)

cv2.imwrite('Re10.jpg',255*g12[0,:,:,10])  
cv2.imwrite('Re50.jpg',255*g12[0,:,:,50])  
cv2.imwrite('Re90.jpg',255*g12[0,:,:,90]) 
````
